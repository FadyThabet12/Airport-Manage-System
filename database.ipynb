```python
import sqlite3
```

```python
connection=sqlite3.connect('Airport Managment System.db')
cursor=connection.cursor()
```

## creat passenger table

```python
cursor.execute('''
create table if not exists Passenger(
               passengerID integer primary key,
               passengerName varchar(14),
               passportNumber varchar(14),
               phoneNumber varchar(11),
               email varchar(30)
)
''')
connection.commit()
```

## staff table

```python
cursor.execute('''
create table if not exists Staff(
                staffID integer primary key,
                staffName varchar(14),
                position varchar(14),
                department varchar(14)
)
''')
connection.commit()
```

## Airline table

```python
cursor.execute('''
create table if not exists Airline(
               AirlineID integer primary key,
                AirlineName varchar(14),
               AirlineCode varchar(14)
)










''')
connection.commit()
```

## gate table

```python
cursor.execute('''
create table if not exists Gate(
                gateID integer primary key,
                gateNumber varchar(14),
                terminal varchar(14)
    )



''')
connection.commit()
```

## Aircraft table

```python
cursor.execute('''
create table if not exists Aircraft(
               AircraftID integer primary key,
               model varchar(50),
               capacity integer,
               AirlineID integer,
               foreign key (AirlineID) references Airline(AirlineID)
)
''')
connection.commit()
```

## Fight table

```python
cursor.execute('''
create table if not exists Flight(
               flightID integer primary key,
               flightNumber varchar(50),
               departureTime datetime,
               arrivalTime datetime,
               origin varchar(50),
               destination varchar(50),
               AircraftID integer,
               gateID integer,
               status varchar(50),
               staffID integer,
               foreign key (AircraftID) references Aircraft(AircraftID),
               foreign key (gateID) references Gate(gateID),
               foreign key (staffID) references Staff(staffID)
)


''')
connection.commit()
```

## seat table

```python
cursor.execute('''
create table if not exists seat(
               seatID integer primary key,
               seatNumber varchar(10),
               class varchar(50),
               status varchar(50),
               AircraftID integer,
                foreign key (AircraftID) references Aircraft(AircraftID)
)
''')
connection.commit()
```

## Booking table

```python
cursor.execute('''
create table if not exists Booking(
                bookingID integer primary key,
                passengerID integer,
                flightID integer,
                seatID integer,
                bookingDate datetime,
                bookingStatus varchar(50),
                foreign key (passengerID) references Passenger(passengerID),
                foreign key (flightID) references Flight(flightID),
                foreign key (seatID) references seat(seatID)
    )


''')
connection.commit()
```

## Baggage

```python
cursor.execute('''
create table if not exists Baggage(
                baggageID integer primary key,
                passengerID integer,
                flightID integer,
                weight decimal(10,2),
                status varchar(50),
                foreign key (passengerID) references Passenger(passengerID),
                foreign key (flightID) references Flight(flightID)
    )


''')
connection.commit()
```

## payment table


```python
cursor.execute('''
create table if not exists payment(
                paymentID integer primary key,
                bookingID integer,
                amount decimal(10,2),
                paymentDate datetime,
                paymentMethod varchar(50),
                paymentStatus varchar(50),
                foreign key (bookingID) references Booking(bookingID)
    )


''')
connection.commit()
```
